## 1. Energiemarkt

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **Day-Ahead-Markt** | Stromhandelsmarkt, an dem am Vortag Preise für jede Stunde des nächsten Tages festgelegt werden. ENTSO-E publiziert die Resultate täglich um ~12:00 Uhr. | NB01, NB02, NB08 |
| **Spot-Preis** | Der an der Strombörse (hier: EPEX SPOT) festgelegte stündliche Preis für Strom [EUR/MWh]. Synonym für Day-Ahead-Preis im Kontext dieses Projekts. | alle |
| **Intraday-Markt** | Handel in Echtzeit innerhalb des laufenden Tages — Preise weichen vom Day-Ahead ab. Nicht Gegenstand des Pflichtteils, relevant für NB10. | NB10 |
| **Spread** | Preisspanne zwischen teuren und günstigen Stunden eines Tages [EUR/MWh]. Direkter Erlöstreiber der Grid-Arbitrage: je grösser der Spread, desto mehr verdient die Batterie. | NB02, NB08 |
| **Negativpreis** | Stunden, in denen Strom einen negativen Preis hat [EUR/MWh < 0]. Entsteht bei Überproduktion (z. B. Sonnentag mit viel Solar). Ideales Ladefenster für Batterien. | NB02, NB03 |
| **Netto-Export** | Differenz Exporte − Importe über alle Grenzen [MW]. Positiv = CH exportiert, negativ = CH importiert. | NB07 |
| **Regelblock** | Geographische Zone der Frequenzregelung. Die Schweiz bildet einen eigenen Regelblock, gesteuert durch Swissgrid. | NB06 |
| **FCR** | *Frequency Containment Reserve* — Primärreserve. Automatische Reaktion auf Frequenzabweichungen innerhalb von Sekunden. In der Schweiz über Swissgrid ausgeschrieben. | NB09 |
| **aFRR** | *Automatic Frequency Restoration Reserve* — Sekundärreserve. Koordinierter Dispatch auf Signal von Swissgrid innerhalb von Minuten. Für aggregierte Batterien zugänglich. | NB09 |
| **VPP** | *Virtual Power Plant* — Virtuelles Kraftwerk. Aggregation vieler kleiner Flexibilitätsressourcen (z. B. Heimspeicher) zu einer koordinierten steuerbaren Einheit. | NB05, NB09 |
| **Kapazitätsprämie** | Vergütung für das Bereithalten von Leistung — unabhängig davon ob diese tatsächlich abgerufen wird [EUR/kW/Jahr]. Relevanter Zusatzerlös bei VPP-Teilnahme. | NB09 |
| **Smart Tariff** | Netzentgelt-Modell mit variablen Tarifen je nach Tageszeit und Netzauslastung. Koordinierte Batterien können Netzentgelt reduzieren. | NB09 |
| **ENTSO-E** | *European Network of Transmission System Operators for Electricity*. Europäischer Dachverband der Übertragungsnetzbetreiber, betreibt die Transparency Platform (Datenquelle für DS1–DS3). | NB01 |
| **Swissgrid** | Schweizerische Übertragungsnetzbetreiberin. Zuständig für Netzstabilität, Regelenergie-Ausschreibungen und (zukünftig) den Flexibilitätsmarkt. | NB05, NB06, NB09 |
| **BFE** | *Bundesamt für Energie*. Betreibt das GeoPackage der Elektrizitätsproduktionsanlagen (DS3, ~322k Anlagen). | NB01, NB06 |
| **BFS** | *Bundesamt für Statistik*. Liefert Kantonsbevölkerungsdaten via PXWeb API (STATPOP). | NB06 |
| **Eigenverbrauchsoptimierung** | Nutzung einer Batterie (aus dem Netz günstig geladen) für den eigenen Strombedarf statt Einspeisung ins Netz. Kein Arbitrage-Erlös, sondern Kostensenkung durch Tarif-Differenz HT/NT. | NB13 |
| **HT / NT** | *Hoch- und Niedertarif* — Zeitabhängiges Netzentgelt-Modell. HT = tagsüber teurer (08–20 Uhr), NT = nachts/Wochenende günstiger. Tarif-Differenz CH ~0.058–0.097 EUR/kWh (0.06–0.10 CHF × 0.97 EUR/CHF). | NB13 |

## 2. Finanzen & Wirtschaftlichkeit

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **Arbitrage** | Im Finanzsinne: risikofreier Gewinn durch Ausnutzung von Preisdifferenzen. Im Energiekontext: *zeitliche* Arbitrage durch Kaufen bei niedrigen und Verkaufen bei hohen Preisen. Nicht vollständig risikofrei (Preisprognose-Risiko, Degradation). | alle |
| **Grid-Arbitrage** | Speziell: Strom-Arbitrage über das öffentliche Netz. Batterie lädt bei niedrigen Spot-Preisen und speist bei hohen Preisen ein. | alle |
| **CAPEX** | *Capital Expenditure* — Investitionskosten. Einmalige Anschaffungskosten der Batterie inkl. Wechselrichter und Installation [EUR oder EUR/kWh]. | NB02, NB04 |
| **OPEX** | *Operating Expenditure* — Betriebskosten. Laufende jährliche Kosten (Wartung, Versicherung), hier angesetzt als 1.5% des CAPEX/Jahr. | NB02, NB04 |
| **ROI** | *Return on Investment* — Rendite. Hier: Netto-Jahreserlös / CAPEX [%/Jahr]. Zielwert: 5% p.a. | NB02, NB04 |
| **Amortisation** | Zeitraum bis das investierte Kapital durch Erlöse zurückgeflossen ist [Jahre]. Synonym: Payback Period, Break-Even-Zeitpunkt. | NB02, NB03 |
| **Break-Even** | Der Zeitpunkt, an dem der kumulierte Netto-Cashflow null ergibt — Investition ist zurückgeflossen. | NB03, NB04 |
| **Cashflow** | Netto-Geldfluss pro Periode. Hier: Einspeisenerlös minus Ladekosten minus OPEX [EUR/Jahr]. | NB02, NB03 |
| **Erlösstacking** | Kombination mehrerer Erlösquellen gleichzeitig: Arbitrage + FCR + aFRR + VPP-Prämie + Smart Tariff. Erhöht den Gesamt-Jahreserlös erheblich. | NB09 |
| **Payback Period** | Synonym für Amortisationsdauer [Jahre]. | NB02, NB04 |
| **Kapazitätsfaktor (CF)** | Verhältnis mittlerer Einspeisung zu installierter Nennleistung [0–1]. Relevant für NB06: CF von Solar ~0.12, Wasser ~0.45, Kernkraft ~0.85. | NB06 |
| **Lernkurve (CAPEX)** | Empirisch beobachteter Rückgang der Batteriekosten mit steigender kumulierter Produktionsmenge. Li-Ion: ca. −10%/Jahr seit 2015. | NB08 |
| **Erlösstacking** | Kombination mehrerer Erlösquellen gleichzeitig: Arbitrage + FCR + aFRR + VPP-Prämie + Smart Tariff. | NB09 |

## 3. Batterie & Technik

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **SoC** | *State of Charge* — Ladezustand der Batterie in Prozent [%]. SoC = 0% = leer, 100% = voll. Im Modell: min. 5% (Tiefentladeschutz), max. 95% (Überladesch.). | NB02, NB04 |
| **Round-Trip-Wirkungsgrad** | Verhältnis entnommener zu eingespeister Energie [%]. Bei Li-Ion typisch 90–95%. Im Projekt: 92% (konservativer Wert). Verlust entsteht durch Lade- und Entladeverluste. | NB02, NB04 |
| **Li-Ion** | *Lithium-Ionen-Technologie*. Dominanter Batterietyp im Heimspeicher- und Gewerbemarkt. Gut dokumentierte Kennwertbasis, daher einzige modellierte Technologie. | NB02, NB04 |
| **Dispatch** | Steuerungsentscheidung: Wann lädt die Batterie, wann speist sie ein, wann bleibt sie idle? Im Modell tagesbasiert nach p25/p75-Schwellenwert. | NB02, NB10 |
| **DA-optimaler Dispatcher** | Dispatch-Strategie basierend auf den am Vortag bekannten Day-Ahead-Preisen. Realistischer als reaktives Modell, da DA-Preise tatsächlich vorab verfügbar sind. | NB10 |
| **Degradation** | Kapazitätsverlust der Batterie über die Lebensdauer [%/Zyklus oder %/Jahr]. Im Modell nicht explizit berücksichtigt (Limitation). | NB04 |
| **GeoPackage (GPKG)** | Offenes Dateiformat für Geodaten (OGC-Standard). BFE Produktionsanlagen werden als GPKG in EPSG:2056 (Schweizer Koordinatensystem) geliefert. | NB06 |
| **EPSG:2056** | Schweizer Landeskoordinatensystem (LV95). Koordinateneinheit: Meter, Ursprung Bern. Wird für NB06-Karten in WGS84 (Länge/Breite) konvertiert. | NB06 |
| **NMC** | *Nickel-Mangan-Kobalt* — Li-Ion-Zellchemie. Höhere Energiedichte als LFP, aber erhöhtes Brandrisiko. Dominant in Elektrofahrzeugen; im stationären Bereich zunehmend von LFP verdrängt. | NB11 |
| **Redox-Flow** | *Vanadium-Redox-Flow-Batterie (VRF)*. Energiespeicher trennt Leistung und Kapazität entkoppelbar — ideal für Utility-Grossspeicher. Sehr lange Lebensdauer (> 20 000 Zyklen), niedrigerer Wirkungsgrad als Li-Ion. | NB11 |
| **NaS** | *Natrium-Schwefel-Batterie*. Betriebstemperatur 300–350 °C — für dezentrale CH-Anlagen ungeeignet. Relevant für Utility-Wärmeanwendungen in wärmeren Klimazonen. | NB11 |
| **CAES** | *Compressed Air Energy Storage* — Druckluftspeicher. Überschussstrom komprimiert Luft in Kavernen; Freisetzung durch Turbinen. A-CAES (adiabat): bis 70 % RTE. In CH geographisch kaum umsetzbar. | NB12 |
| **PHS** | *Pumped Hydro Storage* — Pumpspeicher. Grösster stationärer Energiespeicher der Schweiz (~9.5 GW). Lange Lebensdauer, niedrigster CAPEX/kWh, geographisch gebunden. | NB12 |
| **Power-to-X** | Umwandlung von Überschussstrom in andere Energieträger (H₂, CH₄, Wärme). Sehr niedriger Gesamtwirkungsgrad (~30 % Strom-zu-Strom). Für saisonale Langzeitspeicherung und Sektorkopplung relevant — nicht für tägliche Arbitrage. | NB12 |

## 4. Statistik & Modell

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **Perzentil / Quantil** | Schwellenwert, unter dem ein bestimmter Prozentsatz der Datenpunkte liegt. p25 = 25. Perzentil: 25% der Stunden sind günstiger. p75 = 75. Perzentil: 75% der Stunden sind günstiger (= nur 25% teurer). | NB02, NB04 |
| **p25 / p75** | Kurzschreibweise für das 25. bzw. 75. Perzentil der Tagespreise. Ladeentscheidung: Preis < p25 → laden. Einspeisung: Preis > p75 → einspeisen. | NB02, NB03 |
| **Interpolation** | Auffüllen fehlender Stunden durch Schätzung aus Nachbarwerten. Im Projekt: lineare Interpolation, max. 3h Lücke. | NB02 |
| **Clipping** | Begrenzen von Werten auf einen definierten Bereich. Spot-Preise werden auf −500 bis +3000 EUR/MWh geclippt (physikalisch plausible Grenzen). | NB02 |
| **Symlog** | Achsenskalierung: linear in einem Bereich um 0, logarithmisch ausserhalb. Verwendet für Amortisations-Charts wo Werte von negativ bis sehr positiv reichen. | NB03 |
| **Gauss-Modell** | Statistisches Tagesprofilmodell für simulierte Preise: Summe von Gausskurven für Morgen- und Abendspitzen, Mittagstief, plus Rauschen. | NB01b |
| **Volatilität** | Standardabweichung der stündlichen Spot-Preise [EUR/MWh]. Hohe Volatilität = hoher möglicher Spread = besseres Arbitrage-Umfeld. | NB02, NB08 |
| **Gleichzeitigkeitsrate** | Anteil der Batterien, die in derselben Stunde dispatchen [0–1]. Pessimistisch: 15%, Realistisch: 40%, Optimistisch: 70%. Bestimmt die systemische Netzentlastung. | NB02, NB05 |

## 5. Projektspezifische Begriffe

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **BVI** | *Battery Value Index* — projektspezifischer Wertindex je Netzzone. Kombiniert Zonenimbalance und Engpassnähe zu einem standortspezifischen Kennwert. Hoher BVI = guter Batteriestandort. | NB06 |
| **Zonenimbalance** | Differenz zwischen mittlerer Einspeisung und Verbrauch pro Zone [MW]. Defizit-Zonen (Verbrauch > Einspeisung) profitieren stärker von lokalen Batteriespeichern. | NB06 |
| **Erlösstacking** | Kombination mehrerer Erlösquellen für dieselbe Batterie gleichzeitig (Arbitrage + FCR + aFRR + VPP). Erhöht den Jahreserlös von ~50–100 EUR auf ~400–600 EUR für eine 10-kWh-Batterie. | NB09 |
| **Monitoring-Trigger** | Im Projekt definierte Schwellenwerte, bei deren Erreichen eine Investitionsentscheidung empfohlen wird (z. B. CH-Spread dauerhaft > 30 EUR/MWh). | NB05, NB08 |
| **dataindex.csv** | Projektinternes Datenquellen-Register. Jede geladene oder erzeugte Datei wird mit Zeitstempel, Quelle und Status geloggt. Status: `active`, `superseded`, `error`, `partial`. | NB01, NB99 |
| **MODE** | Konfigurationsschalter (`config.json`): `"data"` = echte ENTSO-E/BFE-Daten, `"sim"` = simulierte Testdaten (NB01b/NB02b). | alle |
| **DIR_INTER_SZ** | Szenario-abhängiger Zwischenordner: `data/intermediate/<gleichzeitigkeit>/`. Wirtschaftlichkeit und Szenarien-CSVs landen je nach aktivem Szenario in einem Unterordner. | NB02, NB05 |
