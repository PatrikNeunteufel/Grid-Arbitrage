## 2. Business Case

Grid-Arbitrage mit Batteriespeichern ermГ¶glicht die wirtschaftliche Optimierung von Strombezug und Einspeisung durch die Nutzung zeitlicher Preisunterschiede. **Forschungsfrage:**
> Unter welchen Bedingungen ist Grid-Arbitrage mit Batteriespeichern in der Schweiz wirtschaftlich sinnvoll вҖ” und welchen systemischen Beitrag zur Netzentlastung kГ¶nnen dezentrale Speicher dabei leisten?

> **Hinweis:** Der Pflichtteil (NB01вҖ“NB04) beantwortet den ersten Teil quantitativ: rentabel oder nicht, und warum. Die vollstГӨndige Antwort вҖ” unter welchen Markt-, Technologie- und Regulierungsbedingungen sich das Bild dreht вҖ” liefert [NB05 Business Strategy](05_Business_Strategy.ipynb) als KГјr.

### 4.3 Dispatch-Modell & Bewertungskriterien

Tagesbasiertes Schwellenwertmodell:
- **Laden** wenn Preis < p25 des Tages und SoC < 95%
- **Einspeisen** wenn Preis > p75 des Tages und SoC > 5%
- **Idle** sonst

Bewertungskriterien: JahreserlГ¶s [EUR], ROI [%], Amortisationsdauer [Jahre], Netzentlastung [MW].

### 4.4 Zentrales Modell: Annahmen und Limitationen

**Explizite Annahmen:**

| Parameter | Wert | BegrГјndung |
|-----------|------|------------|
| Round-Trip-Wirkungsgrad | 92% | Konservativer Li-Ion-Wert (Literatur: 90вҖ“95%) |
| CAPEX Privat | 400 EUR/kWh | MarktГјblicher CH Systempreis 2024 (inkl. Wechselrichter) |
| OPEX | 1.5% CAPEX/Jahr | BranchenГјbliche SchГӨtzung Wartung/Versicherung |
| Simulationshorizont | 12 Jahre | Typische Betriebsdauer Li-Ion (80% KapazitГӨtserhalt) |
| Technologie | Li-Ion | Dominanter Typ im CH Heimspeicher- und Gewerbemarkt |

**Modell-Limitationen:**

- Das Dispatch-Modell entscheidet tГӨglich auf Basis von Tages-Quantilen вҖ” dies approximiert einen DA-optimalen Dispatcher (ENTSO-E DA-Preise sind vor Tagesbeginn bekannt), unterschГӨtzt aber mГ¶gliche Intraday-Gewinne
- Keine Modellierung von Degradation Гјber die Betriebsdauer
- Netzentlastungsszenarien basieren auf geschГӨtzten Rollout-Zahlen, keine offiziellen BFE-Ziele
- Keine BerГјcksichtigung von Steuern, FГ¶rderbeitrГӨgen oder Netztarifen

### 4.5 Ergebnisoffenheit

Ein negatives Ergebnis вҖ” Grid-Arbitrage ist bei aktuellen CH Day-Ahead-Preisen nicht rentabel вҖ” ist kein Projektfehler, sondern ein valider Befund. Die KГјr-Notebooks (NB08вҖ“NB10) untersuchen systematisch, unter welchen Bedingungen sich das Bild dreht.
