## Einleitung <a id='einleitung_O_04'></a>

[↑ Inhaltsverzeichnis](#toc_O_04)

Dieses Projekt wurde unter Einsatz von Generativer KI (LLMs) als "Augmented Engineering"-Prozess umgesetzt. Um die akademische Integrität und die fachliche Korrektheit für den geschäftlichen Einsatz sicherzustellen, wurde ein strukturierter Review-Prozess implementiert.

Dieses Protokoll dokumentiert, dass die KI lediglich als **handwerkliches Werkzeug** (Code-Generierung) diente, während die **Architektur, Validierung und finale Entscheidungsgewalt** beim menschlichen Team lagen.

> **Hinweis zur Rückverfolgbarkeit:** Die Iterationen in Abschnitt 3 wurden rückwirkend aus den vollständigen Chat-Protokollen des Claude-Projekts (SC26_Gruppe_2) rekonstruiert. Alle Korrekturen sind durch die Chat-Historie verifizierbar.

### 2.9 Refactoring-Regressions (Folgeprobleme der Patches)

Die Refactoring-Phasen verliefen nicht fehlerfrei. Der folgende Abschnitt dokumentiert die Folgeprobleme, die bei Stichproben durch PN bzw. beim ersten run_all-Lauf aufgefallen sind. Jeder Fall hat ein eigenes Fix-README im Output-Verzeichnis.

Die Fälle sind qualitativ anders als die K-/A-/I-/S-/V-/C-Kategorien: die KI produzierte initial **plausibel wirkende** Patches, die aber strukturelle Schwächen der Patch-Methodik offenbarten (z.B. Regex-Ende-Marker die nicht greifen, Baseline-Auswahl bei kumulativen Patches, Einrückungs-Defaults die in for-Schleifen versagen). Entdeckt wurden sie durch menschliches Testen — **nicht** durch den KI-Output selbst.

| # | Fall | Ursache | Manifestation | Korrektur | Reviewer |
| :--- | :--- | :--- | :--- | :--- | :--- |

```python
import json
import os

CONFIG_FILE = '../sync/config.json'

if os.path.exists(CONFIG_FILE):
    with open(CONFIG_FILE, 'r', encoding='utf-8') as f:
        cfg = json.load(f)

    print('--- MANUELLER PLAUSIBILITAETS-CHECK ---')

    try:
        if 'szenarien' in cfg and 'wirtschaftlichkeit' in cfg['szenarien']:
            wirth = cfg['szenarien']['wirtschaftlichkeit']
        elif 'pflicht' in cfg and 'wirtschaftlichkeit' in cfg['pflicht']:
            wirth = cfg['pflicht']['wirtschaftlichkeit']
        else:
            raise KeyError("Sektion 'wirtschaftlichkeit' weder in 'szenarien' noch in 'pflicht' gefunden.")

        capex_privat = wirth['capex_eur_kwh']['Privat_10kWh']
        opex         = wirth['opex_rate']
        lifetime     = wirth['lifetime_j']
        eur_chf      = cfg.get('eur_chf', 0.97)
        # Abgeleiteter Wert -- nie in JSON gespeichert (Architektur-Konvention)
        ziel_roi     = round(100 / lifetime, 2)

        print(f'Geprueftes Segment : Privat_10kWh')
        print(f'Eingestellter CAPEX: {capex_privat} EUR/kWh')
        print(f'Eingestellter OPEX : {opex*100}% p.a.')
        print(f'Lebensdauer        : {lifetime} Jahre')
        print(f'ZIEL_ROI (abgel.)  : {ziel_roi}% p.a.  <- lokal berechnet, NICHT aus config')
        print(f'Wechselkurs (Fix)  : {eur_chf} EUR/CHF  <- nur informativ')

        print(f'\n[MENSCHLICHER REVIEW-STATUS]')
        if capex_privat == 400:
            print('Status: VALIDATED (Standard-Szenario)')
        elif capex_privat < 200:
            print('Status: OPTIMISTISCH (Entspricht Zielwerten fuer 2026/2030)')
        else:
            print('Status: CUSTOM (Manuell angepasste Marktwerte)')

    except KeyError as e:
        print(f'Strukturfehler in ../sync/config.json: {e}')
        if 'szenarien' in cfg:
            print(f"Unter-Keys in 'szenarien': {list(cfg['szenarien'].keys())}")
else:
    print(f'Fehler: {CONFIG_FILE} nicht gefunden.')
```

### Befund

Die Fehlerverteilung zeigt zwei deutlich unterschiedliche Phasen der KI-Schwäche:

**Phase 1 — inhaltliche Arbeit (K/A/I/S/V/C/H, 37 Iterationen):**  
Die KI ist im Bereich **Architektur und Konventionen** am anfaelligsten -- sie neigt dazu, lokal plausible Loesungen zu generieren, ohne die projektuebergreifenden Konsistenzregeln zu kennen oder zu halten. **Sachfehler** (falsche Fakten in der Prosa) wurden zuverlaessig erst durch menschliche Fachpruefung entdeckt, da die KI intern konsistente aber inhaltlich falsche Texte produzieren kann.

**Phase 2 — Refactoring (R/F, 20 Iterationen):**  
Beim Refactoring generiert die KI initial **plausibel wirkende** Patches, die aber strukturelle Schwächen der Patch-Methodik offenbaren: Regex-Ende-Marker die in bestimmten Zellen nicht greifen (F-01, F-08), Einrückungs-Defaults die in for-Schleifen versagen (F-07), Baseline-Auswahl bei kumulativen Patches (F-02), übersehene Verwendungen beim Löschen von Funktionen (F-01, F-02, F-08). Diese Fehler sind **nicht** durch ein Selbst-Review der KI detektierbar — sie treten erst bei echtem Ausführen des NB durch Menschen oder `run_all` auf.

Besonders hervorzuheben ist die Kategorie **KI-Halluzinationen im eigenen Review** (H-01 bis H-03): Diese Fehler entstanden nicht beim Code-Generieren, sondern beim *Prüfen des eigenen Outputs* -- die KI erfand plausibel klingende Befunde (Funktionsnamen, JSON-Keys, Produzenten), die im tatsaechlichen Code nie existierten. Sie belegen exemplarisch, warum menschliche Verifikation unverzichtbar ist: **das Werkzeug ist auch bei Selbstaussagen ueber das eigene Verhalten nicht zuverlaessig.**

Analog zu den H-Fällen zeigt die Refactoring-Phase: das Tool kann **syntaktisch korrekten** Patch-Output produzieren, der **funktional defekt** ist (z.B. F-01 `draw_base_map`-Verlust bei AST-valider Zelle, F-08 `_render_grid_background`-Verlust mit AST-valider Lib-Migration). AST-Validierung allein ist keine hinreichende Prüfung.

### Prinzipien aus dem Refactoring

Aus den R/F-Fällen haben sich 12 übertragbare Prinzipien destilliert, die im `Refactoring_Plan.md` und in dieser Sektion als Lehre für zukünftige Projekte festgehalten sind (Auszug):

1. Inspektion gegen volle Baseline, nicht gegen engeren Phase-Output
2. Inspektions-Regex mit Kontext-Ausgabe validieren
3. Numerische Verifikation bit-identisch für hoch-Risiko-Patches
4. User-Stand ist Baseline bei bereits editierten Files
5. Eine Zelle = eine Sache (SRP)
6. Pattern-Abweichungen sind Red Flags
7. Inspektions-Pattern breit genug — nicht nur `def fn_name`, auch charakteristische Konstanten
8. Ende-Marker bei Regex-Ersetzungen müssen robust sein (F-01, F-08)
9. Baseline-Auswahl bei kumulativen Patches: letzter konsolidierter Stand statt Ur-Upload (F-02)
10. Einrückung aus Kontext ermitteln — `call_indent + 4` für Argumente, nicht pauschale Defaults (F-07)
11. show_source-Block-Einfügung als fester Teil von Migrations-Scripten (F-03)
12. Vor dem Löschen einer Funktion projektweit nach Verwendungen suchen (F-01, F-08)
