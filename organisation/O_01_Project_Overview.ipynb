## 1. Motivation <a id='motivation_O_01'></a>

[↑ Inhaltsverzeichnis](#toc_O_01)

### Extrinsische Motivation — Leistungsnachweis

Dieses Projekt ist der Leistungsnachweis für den Kurs *Scripting* im Rahmen des CAS Information Engineering an der ZHAW School of Engineering. Die Aufgabe erfordert:

- Eigenständige **Datenbeschaffung** aus offenen Quellen (APIs, Open Data)
- **Datenbereinigung und -analyse** mit Python/Pandas in Jupyter Notebooks
- **Visualisierung** der Ergebnisse mit mindestens zwei Diagrammtypen (inkl. Heatmap)
- Einen **Business Case** als viertes Notebook

### Intrinsische integrale Motivation — Beruflicher Kontext

Für Teile der Gruppe deckt sich die Aufgabenstellung direkt mit dem beruflichen Betätigungsfeld: Leistungselektronik, Batteriespeicher, Energiemärkte und die wirtschaftliche Bewertung dezentraler Flexibilitätsressourcen. Das Projekt erlaubt es, akademische Anforderungen mit echtem analytischen Erkenntnisinteresse zu verbinden.

Konkret: Das Ergebnis dieses Projekts kann als Entscheidungsgrundlage für **reale Investitionsentscheide** in Batteriespeicher-Infrastruktur dienen, oder auch der Produktentwicklung in diesem Segment — insbesondere da der Schweizer Flexibilitätsmarkt (Swissgrid) in den nächsten Jahren geöffnet wird.

## 2. Business Case <a id='business-case_O_01'></a>

[↑ Inhaltsverzeichnis](#toc_O_01)

[Grid-Arbitrage](O_02_Glossar.ipynb#g-grid-arbitrage) mit Batteriespeichern ermöglicht die wirtschaftliche Optimierung von Strombezug und Einspeisung durch die Nutzung zeitlicher Preisunterschiede. In der Schweiz führt der steigende Anteil erneuerbarer Energien zu höherer [Volatilität](O_02_Glossar.ipynb#g-volatilitaet) bei Strompreisen und Netzlast — wodurch sich systematische Potenziale für speicherbasierte Arbitrage ergeben.

**Forschungsfrage:**
> Unter welchen Bedingungen ist Grid-Arbitrage mit Batteriespeichern in der Schweiz wirtschaftlich sinnvoll — und welchen systemischen Beitrag zur [Netzentlastung](O_02_Glossar.ipynb#g-netzentlastung) können dezentrale Speicher dabei leisten?

> **Hinweis:** Der Pflichtteil (NB00–NB04) beantwortet den ersten Teil quantitativ: rentabel oder nicht, und warum. Die vollständige Antwort — unter welchen Markt-, Technologie- und Regulierungsbedingungen sich das Bild dreht — liefert [K00 Business Strategy](../kuer/K_00_Business_Strategy.ipynb) als Kür.

### Stakeholder

| Stakeholder | Primäres Interesse |
|-------------|-------------------|
| Netzbetreiber (Swissgrid) | Netzstabilität, geringerer Ausbaubedarf |
| Privathaushalte | Kostenersparnis beim Strombezug |
| Gewerbe / Industrie | Lastspitzenreduktion und Arbitrage-Erlöse |
| Energieversorger / Investoren | Portfolio-Optimierung |
| Politik / [BFE](O_02_Glossar.ipynb#g-bfe) | Entscheidungsgrundlage für Energiestrategie 2050 |

### 4.3 Dispatch-Modell & Bewertungskriterien

Tagesbasiertes Schwellenwertmodell:
- **Laden** wenn Preis < [p25](O_02_Glossar.ipynb#g-p25p75) des Tages und [SoC](O_02_Glossar.ipynb#g-soc) < 95%
- **Einspeisen** wenn Preis > [p75](O_02_Glossar.ipynb#g-p25p75) des Tages und SoC > 5%
- **Idle** sonst

Bewertungskriterien: Jahreserlös [EUR], [ROI](O_02_Glossar.ipynb#g-roi) [%], Amortisationsdauer [Jahre], [Netzentlastung](O_02_Glossar.ipynb#g-netzentlastung) [MW].

### 4.4 Zentrales Modell: Annahmen und Limitationen

**Explizite Annahmen:**

| Parameter | Wert | Begründung |
|-----------|------|------------|
| [Round-Trip-Wirkungsgrad](O_02_Glossar.ipynb#g-rte) | 92% | Konservativer [Li-Ion](O_02_Glossar.ipynb#g-li-ion)-Wert (Literatur: 90–95%) |
| [CAPEX](O_02_Glossar.ipynb#g-capex) Privat | 400 EUR/kWh | Marktüblicher CH Systempreis 2024 (inkl. Wechselrichter) |
| [OPEX](O_02_Glossar.ipynb#g-opex) | 1.5% CAPEX/Jahr | Branchenübliche Schätzung Wartung/Versicherung |
| Simulationshorizont | 12 Jahre | Typische Betriebsdauer Li-Ion (80% Kapazitätserhalt) |
| Technologie | Li-Ion | Dominanter Typ im CH Heimspeicher- und Gewerbemarkt |

**Modell-Limitationen:**

- Das Dispatch-Modell entscheidet täglich auf Basis von Tages-Quantilen — dies approximiert einen DA-optimalen Dispatcher (ENTSO-E DA-Preise sind vor Tagesbeginn bekannt), unterschätzt aber mögliche Intraday-Gewinne
- Keine Modellierung von [Degradation](O_02_Glossar.ipynb#g-degradation) über die Betriebsdauer
- Netzentlastungsszenarien basieren auf geschätzten Rollout-Zahlen, keine offiziellen BFE-Ziele
- Keine Berücksichtigung von Steuern, Förderbeiträgen oder Netztarifen

### 4.5 Ergebnisoffenheit

Ein negatives Ergebnis — [Grid-Arbitrage](O_02_Glossar.ipynb#g-grid-arbitrage) ist bei aktuellen CH Day-Ahead-Preisen nicht rentabel — ist kein Projektfehler, sondern ein valider Befund. Die Kür-Notebooks (K_03–K_06) untersuchen systematisch, unter welchen Bedingungen sich das Bild dreht.

## 4.6 Konfigurationsdateien:<a id='konfigurationsdateien_O_01'></a>

[↑ Inhaltsverzeichnis](#toc_O_01)

[config.json](../sync/config.json) und [transfer.json](../sync/transfer.json) 

Das Projekt trennt **statische Inputs** (Schalter, Parameter) von **berechneten Outputs** (Simulationsergebnisse) in zwei separate JSON-Dateien.

### config.json — Single Source of Truth für alle Parameter

Wird vom **User** gepflegt. Notebooks lesen nur, schreiben nie.

| Sektion | Schlüssel (Beispiel) | Beschreibung |
|---------|----------------------|--------------|
| `mode` | `"data"` | `"data"` = echte Daten. `"sim"` war für Dev-Notebooks vorgesehen — **nicht mehr unterstützt** (→ O_03 Sektion 12.3) |
| `eur_chf` | `0.97` | Fixkurs EUR→CHF (März 2026). Alle Berechnungen in EUR. |
| `daten` | `start_year`, `end_year` | Datenzeitraum; `"heute"` = aktuelles Jahr automatisch |
| `force_reload` | `spot_prices`, `netzlast`, … | `true` = Datei neu laden auch wenn vorhanden |
| `pflicht.simulation` | `charge_quantile`, `efficiency_roundtrip`, … | [Dispatch](O_02_Glossar.ipynb#g-dispatch)-Modell-Parameter |
| `pflicht.wirtschaftlichkeit` | `capex_eur_kwh`, `opex_rate`, `lifetime_j`, … | Wirtschaftlichkeitsrechnung |
| `szenarien` | `gleichzeitigkeit_aktiv`, `optionen.realistisch.rate`, … | [VPP](O_02_Glossar.ipynb#g-vpp)-Szenarien |
| `kuer` | `markt.capex_ziel_privat_eur_kwh`, … | Parameter ausschliesslich für Kür-NBs |
| `kuer_aktiv` | `K_01: true`, `K_99: true`, … | Einreichungssteuerung Kür-Notebooks |
| `api_keys` | `entsoe` | [API](O_02_Glossar.ipynb#g-api)-Schlüssel — nie im Notebook-Code hardcoden |

```python
# Laden in jedem Notebook (erste Code-Zelle nach Imports):
import json as _json
with open('../sync/config.json') as _f:
    CFG = _json.load(_f)
MODE    = CFG['mode']          # Alias — nur lesend, nie direkt setzen
EUR_CHF = CFG.get('eur_chf', 0.97)
```

### transfer.json — Berechnete Zwischenergebnisse zwischen Notebooks

Wird von **Notebooks** geschrieben. Jedes Notebook überschreibt nur seine eigene Sektion; alle anderen Sektionen bleiben unverändert.

| Sektion | Geschrieben von | Gelesen von | Inhalt |
|---------|-----------------|-------------|--------|
| `datenzeitraum` | NB01 | NB02–K_99 | `start_date`, `end_date`, `n_years` aus echten Preisdaten |
| `simulation` | NB02 | NB04, K_00, K_05, K_10, K_99 | Spread-Kennzahlen, ROI/Payback je Segment |
| `dispatch_optimierung` | K_06 | K_00 | DA-optimal vs. reaktiv je Segment |
| `eigenverbrauch` | K_09 | K_10 | [ROI](O_02_Glossar.ipynb#g-roi), Jahresersparnis Eigenverbrauchsmodell |
| `hybrid_simulation` | K_99 | K_10 | ROI aller 4 Dispatch-Modelle je Segment |
| `produkt` | K_10 | K_00 | Steckbrief-Kennzahlen des konkreten Produkts |

```python
# Sicher laden (robust gegen leere Datei zu Projektbeginn):
import os as _os, json as _json
_tf_path = '../sync/transfer.json'
_tf = _json.loads(open(_tf_path).read() or '{}') \
      if _os.path.exists(_tf_path) and _os.path.getsize(_tf_path) > 0 else {}
```

```python
# Schreiben (nur eigene Sektion — andere bleiben erhalten):
_tf['simulation'] = { ... }                           # nur diese Sektion ändern
with open(_tf_path, 'w') as _f:
    _json.dump(_tf, _f, indent=2, ensure_ascii=False) # Rest bleibt unverändert
```

> **Ausführungsreihenfolge für vollständiges ../sync/transfer.json:**  
> NB01 → NB02 → K_06 → K_09 → K_99 → K_10  
> Ein einzelnes NB kann jederzeit neu ausgeführt werden — es überschreibt nur seine eigene Sektion.

### 6.1 Library-Architektur <a id='library-architektur_O_01'></a>

Die `lib/`-Module sammeln Funktionen die in mindestens zwei Notebooks gleich
verwendet werden. Ziel: SSOT auch für Code — Änderung an einer Stelle statt
Copy-Paste-Synchronisation über viele Notebooks.

Im `Notebook_Dokumentation.md` findest du für jede lib-Funktion Signatur,
Parameter und die Liste aller NBs die sie aufrufen.

| Modul | Funktionen / Symbole | Hauptzweck |
|-------|----------------------|------------|
| `lib.plotting` | `show_source`, `should_skip`, `make_gif_chart`, `show_chart`, `make_spline_h24` | Chart-Anzeige, Quellcode-Inline-Anzeige, Skip-Logik für teure Chart-Erzeugung |
| `lib.widgets` | `slide_or_play`, `show_animation` | Interaktive Anzeige von Animationen (GIF oder Frame-Slider) |
| `lib.io_ops` | `log_dataindex`, `load_transfer`, `save_transfer`, `needs_download`, `needs_rebuild`, `log_missing`, `final_check`, `pipeline_overview` | Datenregister, Transfer-JSON, Reload-Prüfungen, NB-Abschlusskontrolle, Pipeline-Übersicht |
| `lib.data_fetchers` | `fetch_entsoe_yearly` | ENTSO-E API — jahresweiser Download mit Retry-Logik |
| `lib.simulation` | `simulate_battery_dispatch` | Dispatch-Kernalgorithmus (Quantil-basiert, vektorisiert) |
| `lib.columns` | `find_col` | Spaltennamen-Detektion über Synonyme (Backup gegen CSV-Schema-Änderungen) |
| `lib.grid_topo` | `load_kantone`, `KANT_NUM_TO_ABK`, `KANT_ABK_SET`, `KANT_NAME_TO_ABK` | Schweizer Kantons-Geodaten (Download, Kürzel-Mapping, Namens-Mapping DE/FR/IT) |
| `lib.map_animation` | `make_gif_fast_map`, `clear_background_cache` | Performance-optimierter Karten-GIF-Renderer mit Background-Cache |
| `lib.bfe_zonen` | `load_bfe_plants`, `aggregate_zone_prod`, `aggregate_zone_total_inst`, `aggregate_zone_mean_prod` | [BFE-Anlagen-Loader und Zonen-Aggregation](O_02_Glossar.ipynb#g-bfe-zonen-aggregation) — vereinheitlicht Energieträger-Mapping (Solar/Wasserkraft/Kernkraft/Andere) und kanton→zone-Aggregation für K_01b/c/d |

**Wann lib, wann NB-lokal?** Faustregel: Wenn dieselbe Logik in ≥2 NBs
auftauchen würde → lib. Wenn nur ein NB die Funktion braucht und sie mehr
als 20 Zeilen hat → NB-lokal. Diese Regel ist in `O_03_Konventionen.ipynb`
näher ausgeführt.

**Zugriff aus einem Notebook:**

```python
import sys, os
_PROJECT_ROOT = os.path.abspath('..')
if _PROJECT_ROOT not in sys.path:
    sys.path.insert(0, _PROJECT_ROOT)

from lib.plotting import show_source, show_chart
from lib.io_ops   import log_dataindex
# ... weitere Module nach Bedarf
```

`show_source(fn)` zeigt bei erstem Aufruf einer lib-Funktion den Quellcode
aufklappbar inline an — so muss der Review keine separaten Dateien öffnen
um den Code zu prüfen. Konvention dazu in `O_03_Konventionen.ipynb`.

## 7. Bewertung der Datenqualität <a id='bewertung-der-datenqualitaet_O_01'></a>

[↑ Inhaltsverzeichnis](#toc_O_01)

| Kriterium | Massnahme |
|-----------|----------|
| Plausibilität | Wertebereich-Prüfung (z. B. Preise -500 bis +3000 EUR/MWh) |
| Vollständigkeit | Stundenindex erzwingen, Lücken dokumentieren in `data/missing.txt` |
| Konsistenz | [UTC](O_02_Glossar.ipynb#g-utc)-Normierung, Zeitzonensynchronisation aller Quellen |
| Fehlende Werte | Lineare [Interpolation](O_02_Glossar.ipynb#g-interpolation) (max. 3h), Forward/Backward-Fill für längere Lücken |
| Ausreisser | [Clipping](O_02_Glossar.ipynb#g-clipping) auf physikalisch plausible Grenzen |
| **Währung** | Alle Berechnungen in EUR. CH-Tarife (CHF) werden mit `eur_chf` aus `../sync/config.json` umgerechnet. Kurs fix März 2026 (→ Designentscheidung Sektion 4.4). |
| [API](O_02_Glossar.ipynb#g-api)-Besonderheiten | `query_load` gibt DataFrame statt Series, MW→GW-Konvertierung |

Alle Download-Events sind in `../sync/dataindex.csv` protokolliert (status: `active` / `superseded` / `error` / `partial`).
