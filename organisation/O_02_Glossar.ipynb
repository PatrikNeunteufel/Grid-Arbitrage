## 1. Energiemarkt <a id='energiemarkt_O_02'></a>

[↑ Inhaltsverzeichnis](#toc_O_02)

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **<a id="g-day-ahead-markt"></a>Day-Ahead-Markt** | Stromhandelsmarkt, an dem am Vortag Preise für jede Stunde des nächsten Tages festgelegt werden. ENTSO-E publiziert die Resultate täglich um ~12:00 Uhr. | NB01, NB02, K_03 |
| **<a id="g-spot-preis"></a>Spot-Preis** | Der an der Strombörse (hier: EPEX SPOT) festgelegte stündliche Preis für Strom [EUR/MWh]. Synonym für Day-Ahead-Preis im Kontext dieses Projekts. | alle |
| **<a id="g-duck-curve"></a>Duck Curve / Duck-Curve-Effekt** | Charakteristisches Tagesprofil der Netzlast nach Abzug der Solarproduktion: Residuallast sinkt tagsüber stark ab (Bauch der Kurve, Solar drückt Preis), steigt abends steil an (Schwanz der Kurve, Haushalts- und Industrielast). Die Kurve ähnelt einer Ente (englisch: duck). Stark ausgeprägt im Frühling, wenn Solar bereits produziert aber Heizlast noch hoch ist — erzeugt dort den maximalen Intraday-Spread. | NB04, K_00, K_01, K_03 |
| **Intraday-Markt** | Handel in Echtzeit innerhalb des laufenden Tages — Preise weichen vom Day-Ahead ab. Nicht Gegenstand des Pflichtteils, relevant für K_06. | K_06 |
| **<a id="g-marktpreisspanne"></a>Marktpreisspanne (max−min)** | Mittlere tägliche Differenz zwischen Höchst- und Tiefstwert des Spot-Preises pro Tag, gemittelt über die Saison [EUR/MWh]. Theoretische Obergrenze des erzielbaren Arbitrage-Erlöses an einem Durchschnittstag — setzt perfektes Timing voraus. Saisonal (CH-Daten): Frühling ~84, Sommer ~75, Herbst ~70, Winter ~49 EUR/MWh. Wird in NB04 §5 als Orientierungsgrösse ausgewiesen. **Nicht** Basis des Simulations-Triggers. | NB04 |
| **<a id="g-intra-tag-spread"></a>Intra-Tag-Spread (p75−p25)** | Differenz zwischen 75. und 25. Preisperzentil der 24 Tagesstunden [EUR/MWh]. Massgeblicher Erlöstreiber der Simulation: die Dispatch-Logik kauft unter p25 und verkauft über p75 — daher spiegelt dieser Spread die tatsächlich nutzbare Preisspanne wider. CH-Monatsmedian: Ø ~24 EUR/MWh (`spread_mean_eur_mwh` in ../sync/transfer.json, aktueller Stand). Basis für den Rentabilitäts-Trigger: **> 30 EUR/MWh** dauerhaft → Investition wird attraktiv. | NB02, K_03, K_05, K_99 |
| **Negativpreis** | Stunden, in denen Strom einen negativen Preis hat [EUR/MWh < 0]. Entsteht bei Überproduktion (z. B. Sonnentag mit viel Solar). Ideales Ladefenster für Batterien. | NB02, NB03 |
| **Netto-Export** | Differenz Exporte − Importe über alle Grenzen [MW]. Positiv = CH exportiert, negativ = CH importiert. | K_02 |
| **Regelblock** | Geographische Zone der Frequenzregelung. Die Schweiz bildet einen eigenen Regelblock, gesteuert durch Swissgrid. | K_01 |
| **<a id="g-fcr"></a>FCR** | *Frequency Containment Reserve* — Primärreserve. Automatische Reaktion auf Frequenzabweichungen innerhalb von Sekunden. In der Schweiz über Swissgrid ausgeschrieben. | K_05 |
| **<a id="g-afrr"></a>aFRR** | *Automatic Frequency Restoration Reserve* — Sekundärreserve. Koordinierter Dispatch auf Signal von Swissgrid innerhalb von Minuten. Für aggregierte Batterien zugänglich. | K_05 |
| **<a id="g-vpp"></a>VPP** | *Virtual Power Plant* — Virtuelles Kraftwerk. Aggregation vieler kleiner Flexibilitätsressourcen (z. B. Heimspeicher) zu einer koordinierten steuerbaren Einheit. | K_00, K_05 |
| **Kapazitätsprämie** | Vergütung für das Bereithalten von Leistung — unabhängig davon ob diese tatsächlich abgerufen wird [EUR/kW/Jahr]. Relevanter Zusatzerlös bei VPP-Teilnahme. | K_05 |
| **Smart Tariff** | Netzentgelt-Modell mit variablen Tarifen je nach Tageszeit und Netzauslastung. Koordinierte Batterien können Netzentgelt reduzieren. | K_05 |
| **<a id="g-entsoe"></a>ENTSO-E** | *European Network of Transmission System Operators for Electricity*. Europäischer Dachverband der Übertragungsnetzbetreiber, betreibt die Transparency Platform (Datenquelle für DS1–DS3). | NB01 |
| **Swissgrid** | Schweizerische Übertragungsnetzbetreiberin. Zuständig für Netzstabilität, Regelenergie-Ausschreibungen und (zukünftig) den Flexibilitätsmarkt. | K_00, K_01, K_05 |
| **<a id="g-bfe"></a>BFE** | *Bundesamt für Energie*. Betreibt das GeoPackage der Elektrizitätsproduktionsanlagen (DS3, ~322k Anlagen). | NB01, K_01 |
| **<a id="g-bfs"></a>BFS** | *Bundesamt für Statistik*. Liefert Kantonsbevölkerungsdaten via PXWeb API (STATPOP). | K_01 |
| **Eigenverbrauchsoptimierung** | Nutzung einer Batterie (aus dem Netz günstig geladen) für den eigenen Strombedarf statt Einspeisung ins Netz. Kein Arbitrage-Erlös, sondern Kostensenkung durch Tarif-Differenz HT/NT. | K_09 |
| **HT / NT** | *Hoch- und Niedertarif* — Zeitabhängiges Netzentgelt-Modell. HT = tagsüber teurer (08–20 Uhr), NT = nachts/Wochenende günstiger. Tarif-Differenz CH ~0.058–0.097 EUR/kWh (0.06–0.10 CHF × 0.97 EUR/CHF). | K_09 |
| **Merit-Order-Prinzip** | Marktwirtschaftliches Prinzip: Kraftwerke werden in aufsteigender Reihenfolge ihrer Grenzkosten eingesetzt. Günstige Quellen (Solar, Wind, Kernkraft) kommen zuerst; teure (Erdgas, Öl) setzen den Marginalpreis. CH importiert genau dann, wenn der Marginalpreis im Ausland unter dem Inlandspreis liegt — daher korrelieren hohe Preise mit Import-Stunden. | K_02 |
| **Physikalische Grenzflüsse** | Gemessene Stromflüsse an den Landesgrenzen CH↔DE/AT/IT/FR [MW]. Positiv = Import nach CH, negativ = Export aus CH. Quelle: ENTSO-E Transparency Platform (`query_crossborder_flows`). Empirischer Nachweis: CH importiert in Hochpreis-Stunden, exportiert in Niedrigpreis-Stunden. | K_02 |

| **Arbitrage** | Im Finanzsinne: risikofreier Gewinn durch Ausnutzung von Preisdifferenzen. Im Energiekontext: *zeitliche* Arbitrage durch Kaufen bei niedrigen und Verkaufen bei hohen Preisen. Nicht vollständig risikofrei (Preisprognose-Risiko, Degradation). | alle |
| **<a id="g-grid-arbitrage"></a>Grid-Arbitrage** | Speziell: Strom-Arbitrage über das öffentliche Netz. Batterie lädt bei niedrigen Spot-Preisen und speist bei hohen Preisen ein. | alle |

| **<a id="g-opex"></a>OPEX** | *Operating Expenditure* — Betriebskosten. Laufende jährliche Kosten (Wartung, Versicherung), hier angesetzt als 1.5% des CAPEX/Jahr. | NB02, NB04 |
| **<a id="g-roi"></a>ROI** | *Return on Investment* — Rendite. Hier: Netto-Jahreserlös / CAPEX [%/Jahr]. Zielwert: 5% p.a. | NB02, NB04 |
| **Amortisation** | Zeitraum bis das investierte Kapital durch Erlöse zurückgeflossen ist [Jahre]. Synonym: Payback Period, Break-Even-Zeitpunkt. | NB02, NB03 |

| **Cashflow** | Netto-Geldfluss pro Periode. Hier: Einspeisenerlös minus Ladekosten minus OPEX [EUR/Jahr]. | NB02, NB03 |
| **<a id="g-erloess-stacking"></a>Erlösstacking** | Kombination mehrerer Erlösquellen gleichzeitig: Arbitrage + FCR + aFRR + VPP-Prämie + Smart Tariff. Erhöht den Gesamt-Jahreserlös erheblich. | K_05 |
| **Payback Period** | Synonym für Amortisationsdauer [Jahre]. | NB02, NB04 |
| **Kapazitätsfaktor (CF)** | Verhältnis mittlerer Einspeisung zu installierter Nennleistung [0–1]. Relevant für K_01: CF von Solar ~0.12, Wasser ~0.45, Kernkraft ~0.85. | K_01 |
| **Lernkurve (CAPEX)** | Empirisch beobachteter Rückgang der Batteriekosten mit steigender kumulierter Produktionsmenge. Li-Ion: ca. −10%/Jahr seit 2015. | K_03 |
| **Erlösstacking** | Kombination mehrerer Erlösquellen gleichzeitig: Arbitrage + FCR + aFRR + VPP-Prämie + Smart Tariff. | K_05 |

## 3. Batterie & Technik <a id='batterie-technik_O_02'></a>

[↑ Inhaltsverzeichnis](#toc_O_02)

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **<a id="g-soc"></a>SoC** | *State of Charge* — Ladezustand der Batterie in Prozent [%]. SoC = 0% = leer, 100% = voll. Im Modell: min. 5% (Tiefentladeschutz), max. 95% (Überladesch.). | NB02, NB04 |
| **<a id="g-rte"></a>Round-Trip-Wirkungsgrad** | Verhältnis entnommener zu eingespeister Energie [%]. Bei Li-Ion typisch 90–95%. Im Projekt: 92% (konservativer Wert). Verlust entsteht durch Lade- und Entladeverluste. | NB02, NB04 |
| **<a id="g-li-ion"></a>Li-Ion** | *Lithium-Ionen-Technologie*. Dominanter Batterietyp im Heimspeicher- und Gewerbemarkt. Gut dokumentierte Kennwertbasis, daher einzige modellierte Technologie. | NB02, NB04 |
| **<a id="g-dispatch"></a>Dispatch** | Steuerungsentscheidung: Wann lädt die Batterie, wann speist sie ein, wann bleibt sie idle? Im Modell tagesbasiert nach p25/p75-Schwellenwert. | NB02, K_06 |
| **<a id="g-da-dispatcher"></a>DA-optimaler Dispatcher** | Dispatch-Strategie basierend auf den am Vortag bekannten Day-Ahead-Preisen. Realistischer als reaktives Modell, da DA-Preise tatsächlich vorab verfügbar sind. | K_06 |
| **<a id="g-c-rate"></a>C-Rate** | Verhältnis von Ladeleistung [kW] zu Kapazität [kWh]. C-Rate = 1C bedeutet: volle Kapazität in 1 Stunde lad- oder entladbar. Beispiel: 5 kW / 10 kWh = 0.5C → nur 50% der Kapazität pro Stunde nutzbar. Limitiert den erzielbaren Erlös bei kurzem Preisspread-Fenster. | NB04, K_06 |
| **<a id="g-degradation"></a>Degradation** | Kapazitätsverlust der Batterie über die Lebensdauer [%/Zyklus oder %/Jahr]. Im Modell nicht explizit berücksichtigt (Limitation). | NB04 |
| **GeoPackage (GPKG)** | Offenes Dateiformat für Geodaten (OGC-Standard). BFE Produktionsanlagen werden als GPKG in EPSG:2056 (Schweizer Koordinatensystem) geliefert. | K_01 |
| **<a id="g-epsg2056"></a>EPSG:2056** | Schweizer Landeskoordinatensystem (LV95). Koordinateneinheit: Meter, Ursprung Bern. Wird für K_01-Karten in WGS84 (Länge/Breite) konvertiert. | K_01 |
| **<a id="g-nmc"></a>NMC** | *Nickel-Mangan-Kobalt* — Li-Ion-Zellchemie. Höhere Energiedichte als LFP, aber erhöhtes Brandrisiko. Dominant in Elektrofahrzeugen; im stationären Bereich zunehmend von LFP verdrängt. | K_07 |
| **<a id="g-redox-flow"></a>Redox-Flow** | *Vanadium-Redox-Flow-Batterie (VRF)*. Energiespeicher trennt Leistung und Kapazität entkoppelbar — ideal für Utility-Grossspeicher. Sehr lange Lebensdauer (> 20 000 Zyklen), niedrigerer Wirkungsgrad als Li-Ion. | K_07 |
| **<a id="g-nas"></a>NaS** | *Natrium-Schwefel-Batterie*. Betriebstemperatur 300–350 °C — für dezentrale CH-Anlagen ungeeignet. Relevant für Utility-Wärmeanwendungen in wärmeren Klimazonen. | K_07 |
| **<a id="g-caes"></a>CAES** | *Compressed Air Energy Storage* — Druckluftspeicher. Überschussstrom komprimiert Luft in Kavernen; Freisetzung durch Turbinen. A-CAES (adiabat): bis 70 % RTE. In CH geographisch kaum umsetzbar. | K_08 |
| **<a id="g-phs"></a>PHS** | *Pumped Hydro Storage* — Pumpspeicher. Grösster stationärer Energiespeicher der Schweiz (~9.5 GW). Lange Lebensdauer, niedrigster CAPEX/kWh, geographisch gebunden. | K_08 |
| **<a id="g-power-to-x"></a>Power-to-X** | Umwandlung von Überschussstrom in andere Energieträger (H₂, CH₄, Wärme). Sehr niedriger Gesamtwirkungsgrad (~30 % Strom-zu-Strom). Für saisonale Langzeitspeicherung und Sektorkopplung relevant — nicht für tägliche Arbitrage. | K_08 |
| **<a id="g-netzentlastung"></a>Netzentlastung** | Reduktion der Spitzenlast im Verteil- oder Übertragungsnetz durch lokal gespeicherte und bedarfsgerecht eingespeiste Energie. Batterien in Engpasszonen haben einen höheren Netzwert als in ausgeglichenen Zonen — Grundlage des BVI. | K_01 |
| **<a id="g-engpasszone"></a>Engpasszone** | Netzregion, in der die Übertragungskapazität zwischen Produktions- und Verbrauchszentren physikalisch begrenzt ist. Swissgrid definiert in CH fünf Zonen; Göschenen–Airolo ist der kritischste Engpass (Nord-Süd-Achse). Batterien in Engpasszonen erhalten einen Multiplikator auf ihren BVI. | K_01 |

## 4. Statistik & Modell <a id='statistik-modell_O_02'></a>

[↑ Inhaltsverzeichnis](#toc_O_02)

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **<a id="g-perzentil"></a>Perzentil / Quantil** | Schwellenwert, unter dem ein bestimmter Prozentsatz der Datenpunkte liegt. p25 = 25. Perzentil: 25% der Stunden sind günstiger. p75 = 75. Perzentil: 75% der Stunden sind günstiger (= nur 25% teurer). | NB02, NB04 |
| **<a id="g-p25p75"></a>p25 / p75** | Kurzschreibweise für das 25. bzw. 75. Perzentil der Tagespreise. Ladeentscheidung: Preis < p25 → laden. Einspeisung: Preis > p75 → einspeisen. | NB02, NB03 |
| **<a id="g-dispatch-breakeven"></a>Dispatch-Break-even-Bedingung** | Mindestbedingung damit ein Dispatch-Tag profitabel ist: `p(discharge_q) × η > p(charge_q)`. Ist diese Bedingung nicht erfüllt, übersteigen die Rundlaufverluste den erzielbaren Erlös — der Tag bleibt **idle**. Äquivalente Schreibweise: Spread > Mindest-Spread. | NB03 |
| **<a id="g-mindest-spread"></a>Mindest-Spread** | Minimaler Spread, ab dem ein Dispatch-Tag die Rundlaufverluste deckt: `Spread_min = p(charge_q) × (1/η − 1)`. Bei η = 0.92 und p25 = 25 EUR/MWh ergibt sich Spread_min ≈ 2.2 EUR/MWh. Tage mit kleinerem Spread bleiben idle (kein Laden/Entladen). | NB03 |
| **<a id="g-rel-effizienzverlust"></a>Relativer Effizienzverlust** | Kennzahl `(1/η − 1)`: Anteil der Ladeenergie, der durch Rundlaufverluste verloren geht. Bei η = 0.92: `1/0.92 − 1 ≈ 0.087` → 8.7 % Verlust. Multipliziert mit dem Ladepreis ergibt sich der Mindest-Spread in EUR/MWh. | NB03 |
| **<a id="g-interpolation"></a>Interpolation** | Auffüllen fehlender Stunden durch Schätzung aus Nachbarwerten. Im Projekt: lineare Interpolation, max. 3h Lücke. | NB02 |
| **<a id="g-clipping"></a>Clipping** | Begrenzen von Werten auf einen definierten Bereich. Spot-Preise werden auf −500 bis +3000 EUR/MWh geclippt (physikalisch plausible Grenzen). | NB02 |
| **<a id="g-symlog"></a>Symlog** | Achsenskalierung: linear in einem Bereich um 0, logarithmisch ausserhalb. Verwendet für Amortisations-Charts wo Werte von negativ bis sehr positiv reichen. | NB03 |
| **Gauss-Modell** | Statistisches Tagesprofilmodell für simulierte Preise: Summe von Gausskurven für Morgen- und Abendspitzen, Mittagstief, plus Rauschen. | NB01b |
| **<a id="g-volatilitaet"></a>Volatilität** | Standardabweichung der stündlichen Spot-Preise [EUR/MWh]. Hohe Volatilität = hoher möglicher Spread = besseres Arbitrage-Umfeld. | NB02, K_03 |
| **<a id="g-gleichzeitigkeitsrate"></a>Gleichzeitigkeitsrate** | Anteil der Batterien, die in derselben Stunde dispatchen [0–1]. Pessimistisch: 15%, Realistisch: 40%, Optimistisch: 70%. Bestimmt die systemische Netzentlastung. | NB02, K_00 |

## 5. Projektspezifische Begriffe <a id='projektspezifische-begriffe_O_02'></a>

[↑ Inhaltsverzeichnis](#toc_O_02)

| Begriff | Erklärung | Wo im Projekt |
|---------|-----------|---------------|
| **<a id="g-bvi"></a>BVI** | *Battery Value Index* — projektspezifischer Wertindex je Netzzone. Kombiniert Zonenimbalance und Engpassnähe zu einem standortspezifischen Kennwert. Hoher BVI = guter Batteriestandort. | K_01 |
| **<a id="g-zonenimbalance"></a>Zonenimbalance** | Differenz zwischen mittlerer Einspeisung und Verbrauch pro Zone [MW]. Defizit-Zonen (Verbrauch > Einspeisung) profitieren stärker von lokalen Batteriespeichern. | K_01 |
| **Erlösstacking** | Kombination mehrerer Erlösquellen für dieselbe Batterie gleichzeitig (Arbitrage + FCR + aFRR + VPP). Erhöht den Jahreserlös von ~50–100 EUR auf ~400–600 EUR für eine 10-kWh-Batterie. | K_05 |
| **<a id="g-monitoring-trigger"></a>Monitoring-Trigger** | Im Projekt definierte Schwellenwerte, bei deren Erreichen eine Investitionsentscheidung empfohlen wird (z. B. CH-Spread dauerhaft > 30 EUR/MWh). | K_00, K_03 |
| **../sync/dataindex.csv** | Projektinternes Datenquellen-Register. Jede geladene oder erzeugte Datei wird mit Zeitstempel, Quelle und Status geloggt. Status: `active`, `superseded`, `error`, `partial`. | NB01, NB99 |
| **MODE** | Konfigurationsschalter (`../sync/config.json`): `"data"` = echte ENTSO-E/BFE-Daten, `"sim"` = simulierte Testdaten (NB01b/NB02b). | alle |
| **DIR_INTER_SZ** | Szenario-abhängiger Zwischenordner: `data/intermediate/<gleichzeitigkeit>/`. Wirtschaftlichkeit und Szenarien-CSVs landen je nach aktivem Szenario in einem Unterordner. | NB02, K_00 |
| **<a id="g-engpassmultiplikator"></a>Engpassmultiplikator** | Projektspezifischer Faktor im BVI, der den Netzwert einer Batterie in einer Engpasszone erhöht. Wert: 1.0 (ausgeglichene Zone) bis 2.0 (kritische Engpasszone). Formel: `BVI = Zonenimbalance × Engpassmultiplikator`. | K_01 |
| **<a id="g-rollout-szenario"></a>Rollout-Szenario** | Eines von vier projektierten Ausbaupfaden für Batteriespeicher in der CH (Status Quo 2024 / Moderat 2027 / Ambitioniert 2030 / Transformativ 2035). Definiert Anzahl Einheiten je Segment (Privat/Gewerbe/Industrie) und daraus resultierende Netzentlastung [MW]. | K_01 |
