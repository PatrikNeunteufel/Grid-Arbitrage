## Profile: Alles auf einmal<a id='profile_O_00'></a>

[↑ Inhaltsverzeichnis](#toc_O_00)

Wer's eilig hat: diese Zelle installiert alles Nötige für das gesamte
Projekt in einem Aufruf. Pakete, die bereits installiert sind, werden
übersprungen (kein Schaden durch Wiederholung).

## Abschluss<a id='abschluss_O_00'></a>

[↑ Inhaltsverzeichnis](#toc_O_00)

Prüft, dass alle Basis-Imports funktionieren, und zeigt Versionen an.


```python
# ── Verifikation: importiere alle Pakete, zeige Versionen ────────────────────
import importlib

# Pakete, deren Versionen wir anzeigen wollen (nur die wirklich wichtigen)
VERSION_CHECK = [
    ('pandas', 'pandas'),
    ('numpy', 'numpy'),
    ('matplotlib', 'matplotlib'),
    ('scipy', 'scipy'),
    ('geopandas', 'geopandas'),
    ('shapely', 'shapely'),
    ('networkx', 'networkx'),
    ('PIL', 'PIL'),
    ('entsoe', 'entsoe'),
    ('IPython', 'IPython'),
]

print('Paket-Versionen:')
print('-' * 40)
for disp_name, mod_name in VERSION_CHECK:
    try:
        mod = importlib.import_module(mod_name)
        ver = getattr(mod, '__version__', '?')
        print(f'  {disp_name:15s} {ver}')
    except ImportError:
        print(f'  {disp_name:15s} — (nicht installiert)')

print('\n✅ Installer abgeschlossen — Notebooks sind bereit zur Ausführung.')
print('\n→ Weiter mit O_01_Project_Overview.ipynb für den Projekt-Einstieg.')
```
