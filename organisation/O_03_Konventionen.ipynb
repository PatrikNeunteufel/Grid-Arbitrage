## 1. Zellstruktur <a id='zellstruktur_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### 1.1 Grundprinzip

Jede inhaltliche Einheit besteht aus **drei Teilen** in dieser Reihenfolge:

```
[MD]   ## Abschnittstitel + kurze Beschreibung was der folgende Code tut
[Code] Implementierung
[Code] Verifikation (head / print / assert)
```

**Nie** zwei Code-Zellen nacheinander ohne MD-Trenner — jeder Trenner muss
in einem Satz beschreiben, was der folgende Block tut (nicht nur einen Titel nennen).

**Zwischen der Titel-MD-Zelle und der ersten Code-Zelle** steht ebenfalls eine
MD-Zelle, die den Setup-Block beschreibt (z.B. «Imports und Konfiguration laden»).

**Ausnahme Report-Notebooks (NB04, K_00):** Aufeinanderfolgende reine `show_chart()`-Zellen
brauchen keinen MD-Trenner — sie dienen nur der Chart-Darstellung.

### 1.2 Zellgrösse

- Kommentarzeilen (`#`) werden **nicht** gezählt.
- Zusammengehöriger Code (z.B. eine [Dispatch](O_02_Glossar.ipynb#g-dispatch)-Schleife, ein Visualisierungs-Block)
  bleibt zusammen — Aufteilung nur wenn eine **klare logische Trennung** besteht.
- Faustregel: Eine Code-Zelle = eine klar abgrenzbare Aufgabe.

### 1.3 Erste Zelle jedes Notebooks

Immer eine **Titel-Markdown-Zelle** mit:
```markdown
# NB{N} – {Titel}
### CAS Information Engineering – Scripting Project
**Gruppe:** ... | **Datum:** ...

---
Kurze Beschreibung was dieses Notebook tut.
```

### 1.4 Letzte Zellen jedes Notebooks

Immer eine **Abschluss-Code-Zelle** die Ausgabedateien auflistet und validiert,
gefolgt von einer MD-Zelle mit Verweis auf das nächste Notebook.

### 1.5 Kür-Notebooks: Datenladen

Kür-Notebooks laden ihre Datensätze **im ersten Abschnitt des Notebooks, das sie erstmals benötigt** — nicht in NB01.

```
NB01   → lädt nur Pflicht-Datensätze (DS1 Spot-Preise, DS2 Netzlast)
K_01   → lädt BFE GeoPackage + BFS STATPOP + swisstopo in Sektion 1
K_02   → lädt ENTSO-E Grenzflüsse in Sektion 1
NB0x   → jedes weitere Kür-NB lädt seine Datensätze in Sektion 1
```

**Aufbau Sektion 1 in Kür-Notebooks:**
```
[MD]   ## 1. Daten laden
       Hinweis: «Kür-Datensätze werden hier geladen — erstes NB das sie benötigt»
       Tabelle: Datei | Quelle | Pfad | Erstellt in
[Code] Setup (../sync/config.json, Pfade, needs_download-Funktion)
[MD]   ### 1.1 <Datensatzname> — Quelle, Methode, Format, Zweck
[Code] Download mit Retry + log_dataindex
[Code] Verifikation (head, shape, range)
[MD]   ### 1.2 <nächster Datensatz> ...
```

Wenn ein Kür-NB einen Datensatz konsumiert, der von einem früheren Kür-NB geladen wurde,
steht in Sektion 1 nur eine Lese-Zelle (kein Download) mit Verweis auf das ladende NB.

## 2. Markdown-Konventionen <a id='markdown-konventionen_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### Hierarchie

```
# NB-Titel                    ← nur einmal, ganz oben
## 1. Hauptabschnitt          ← Nummeriert, mit führendem --- <a id='hauptabschnitt_O_03'></a>
### 1.1 Unterabschnitt        ← Nummeriert
**Fetttext**                  ← für Schlüsselbegriffe
`code_inline`                 ← für Variablennamen, Dateinamen
> Blockquote                  ← für wichtige Hinweise / Warnungen
```

### Tabellen

Strukturierte Informationen (Datensätze, Segmente, Ergebnisse) immer als MD-Tabelle,
nicht als Code-Ausgabe.

### Keine Wiederholung von Code-Ausgaben in MD

Nicht: *'Der DataFrame hat 17519 Zeilen'* im MD schreiben — das steht im head().
Stattdessen: Interpretation und Konsequenzen beschreiben.

### 2.1 Konfigurationsabhängige Werte im Markdown (⚙ / 📊) <a id='markdown-werte_O_03'></a>

Zahlen in Markdown-Zellen sind statisch — sie spiegeln den Stand zum Zeitpunkt des
letzten Notebook-Runs wider und werden **nicht automatisch aktualisiert**.
Um veraltete Werte nach Konfigurationsänderungen oder Neuberechnungen zu erkennen,
gelten zwei Pflicht-Marker:

| Marker | Bedeutung | Quelle | Wann prüfen? |
|--------|-----------|--------|--------------|
| `⚙` | Wert stammt aus `../sync/config.json` | config.json → Alias in Setup-Zelle | Nach jeder config-Änderung |
| `📊` | Wert ist Simulations- oder Datenergebnis | `../sync/transfer.json` oder Simulationsoutput | Nach Neuberechnung / neuen Daten |

**Regel:** Jede Zahl in Markdown-Fliesstext oder Tabellen, die aus config.json oder
der Simulation stammt, bekommt den Marker **direkt nach der Zahl** — kein Leerzeichen dazwischen.

```markdown
Nutzungsdauer **12⚙ Jahre**. Wirkungsgrad **92%⚙**. Ziel-ROI: **8.33%⚙** p.a. (= 100 / lifetime_j)

Industrie erreicht **2.3%📊** ROI — nächste am Ziel.
Break-Even: **> 20 Jahre📊** für alle Segmente bei aktuellen Annahmen.
```

**Datenfluss-Architektur** (verbindlich):

```
config.json  →  NB02 berechnet  →  transfer.json  →  alle folgenden NBs laden von transfer
     ↑                                                       ↑
  User schreibt                                        Notebooks schreiben
```

`config.json` enthält ausschliesslich **Parameter** (Schalter, Schwellenwerte, Annahmen).
`transfer.json` enthält ausschliesslich **berechnete Outputs** (Simulation, Kennzahlen).
Markdown-Text zeigt den Stand des letzten Runs — Marker machen veraltete Stellen sichtbar.

**Prüfblock-Pflicht:** Jedes Notebook mit ⚙-Markern im Markdown enthält direkt nach
der Initialisierungs-Setup-Zelle eine Pflicht-Code-Zelle:

```python
# ── ⚙ Markdown-Prüfwerte ──────────────────────────────────────────────────────
# Diese Werte erscheinen als ⚙ im Markdown-Text — nach config-Änderung dort prüfen!
print(f'  ZIEL_ROI   = {ZIEL_ROI_PCT}%  (= 100 / {LIFETIME} Jahre)')
print(f'  EFFICIENCY = {EFFICIENCY*100:.0f}%  |  SOC {SOC_MIN_PCT*100:.0f}%–{SOC_MAX_PCT*100:.0f}%')
print(f'  OPEX_RATE  = {OPEX_RATE*100:.1f}%  |  CAPEX_REF = {CAPEX_EUR_KWH} EUR/kWh')
```

**Warum nicht dynamisch?** Markdown-Zellen sind in Jupyter statischer Text —
eine Code-Zelle kann keine Markdown-Zelle zur Laufzeit befüllen.
Die Marker-Konvention ist der pragmatische Mittelweg: Werte bleiben im Markdown,
werden aber klar als prüfpflichtig ausgewiesen.

## 3. Code-Konventionen <a id='code-konventionen_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### 3.1 Kommentare

```python
# ── Abschnittskommentar ──────────────────────────────────────────────────────
# (80 Zeichen, mit ── und trailing ─)

# Inline-Kommentar: erklärt WARUM, nicht WAS
df['price_eur_mwh'] = df['price_eur_mwh'].clip(-500, 3000)  # physikalisch plausibel
```

### 3.2 Variablennamen

| Typ | Konvention | Beispiel |
|-----|------------|----------|
| DataFrame | `df_` Präfix | `df_prices`, `df_load` |
| Dateipfad | `_FILE` Suffix | `PRICES_FILE`, `CLEAN_FILE` |
| Konstante | UPPER_CASE | `FORCE_RELOAD`, `CHARTS_DIR` |
| Hilfsfunktion | snake_case | `needs_download()`, `log_dataindex()` |
| Plot-Objekte | `fig_`, `ax_` | `fig1, ax1 = plt.subplots(...)` |

### 3.3 Importe

Alle Importe in der **ersten Code-Zelle** des Notebooks.
Keine nachträglichen Importe mitten im Notebook — **Ausnahmen:**

1. **pip-install-Check:** Eine erste Code-Zelle, die fehlende Pakete installiert
   (vor dem eigentlichen Import-Block), ist erlaubt und gilt nicht als Verstoss.
2. **Re-Import mit lokalem Alias:** In isolierten Zellen (z.B. Transfer-Zellen)
   darf ein bereits importiertes Modul mit lokalem Alias re-importiert werden.
   Diese Zeile **muss** mit `# Re-Import mit lokalem Alias (<Zweck>)` kommentiert
   sein, sonst gilt sie als ungekennzeichneter Late-Import → Verstoss.

```python
# ── Importe ──────────────────────────────────────────────────────────────────
import os, warnings
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
# In einer Transfer-Zelle:
import os as _os, json as _json  # Re-Import mit lokalem Alias (Transfer-Zelle)
```

### 3.4 Pfade

Nie hartcodierte absolute Pfade. Immer relativ mit `os.path.join()`:
```python
# Gut
PRICES_FILE = os.path.join(DIR_RAW, 'ch_spot_prices_raw.csv')
# Schlecht
PRICES_FILE = 'C:/Users/patrik/projekt/data/raw/ch_spot_prices_raw.csv'
```

## 4. DataFrame-Verifikation (head-Konvention) <a id='dataframe-verifikation-head-konvention_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### Pflicht: head() nach diesen Operationen

| Auslöser | Verifikations-Code |
|----------|-------------------|
| Laden aus CSV / [API](O_02_Glossar.ipynb#g-api) | `df.head(3)` + shape + dtypes |
| Nach Bereinigung | `df.head(3)` + Nullwert-Check |
| Nach Feature Engineering | `df[neue_spalten].head(3)` |
| Nach Aggregation/Pivot | `df.head(3)` |
| Nach Merge/Join | `df.head(3)` + shape |

### Pflicht-Template für geladene Datensätze

```python
# ── Verifikation ─────────────────────────────────────────────────────────────
print(f'Shape  : {df_prices.shape}')
print(f'Dtypes : {df_prices.dtypes.to_dict()}')
print(f'Nulls  : {df_prices.isnull().sum().sum()}')
print(f'Range  : {df_prices["price_eur_mwh"].min():.1f} – '
      f'{df_prices["price_eur_mwh"].max():.1f} EUR/MWh')
df_prices.head(3)
```

### Wann kein head() nötig

- Nach reinen Plot-Operationen
- Nach `.to_csv()` Aufrufen
- Nach einfachen Spaltenberechnungen ohne neue Struktur
- Innerhalb von Schleifen / Animationsframes

### Tiefe: 3 Zeilen Standard

`df.head(3)` ist Standard. Bei breiten DataFrames (>10 Spalten): `df.head(3).T` für bessere Lesbarkeit.

### 5.1 Zweck — Single Source of Truth

Alle Notebooks lesen ihre Steuerparameter aus `../sync/config.json` im Projektverzeichnis.
**Nie** Schalter im Notebook-Code hardcoden — immer über `../sync/config.json`.


## 6. Datenzeitraum: Empfehlungen & Stolperfallen <a id='datenzeitraum-empfehlungen-stolperfallen_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### Empfehlungen

- **`end_year: 'heute'`** verwenden — so läuft das Skript jedes Jahr neu ohne Änderung
- **Mindestens 2 vollständige Kalenderjahre** laden für saisonale Analyse (NB2, Chart 5)
- **Jährlicher Re-Run** empfohlen: Chart 7 zeigt ob Spread gestiegen/gefallen ist
- Nach Änderung von `start_year`/`end_year`: alle `force_reload`-Keys auf `true` setzen
  und NB1–NB3 komplett neu ausführen (`run_all.sh`)

### Stolperfallen

| Problem | Ursache | Lösung |
|---|---|---|
| [ENTSO-E](O_02_Glossar.ipynb#g-entsoe) gibt 503 zurück | Server temporär überlastet | 15–30 Min warten, dann erneut |
| Lückenhafter Datensatz | Einzelne Jahre nicht verfügbar | In NB1 Abschlusskontrolle prüfen |
| `n_years = 1` statt 2 | `end_year` liegt in der Zukunft | ENTSO-E liefert nur bis heute minus ~2 Tage |
| Simulation gibt `NaN` | Zu wenig Daten für Quantil-Berechnung | Mindestens 3 Monate pro Jahr nötig |
| Spread-Chart springt | Fehlende Monate im Datensatz | `force_reload.clean: true` und NB2 neu |
| Alte CSVs nach Zeitraumänderung | `force_reload` noch auf `false` | Alle relevanten Keys auf `true` |

### Besonderheiten ENTSO-E API

- Day-Ahead-Preise werden typisch **einen Tag nach Handelsschluss** verfügbar
- Für `end_year: 'heute'` wird nur bis zum Vortag geladen — ENTSO-E hat keine Intraday-Daten
- Lücken über **Feiertage** (besonders Ostern, Weihnachten) sind normal und werden interpoliert
- Die ENTSO-E [API](O_02_Glossar.ipynb#g-api) hat **Rate Limits**: zu viele Requests in kurzer Zeit → 429-Fehler
  → jahresweiser Download mit `_fetch_prices_year()` umgeht das Problem
- Wechsel der **Zeitzone** (CET↔CEST) kann zu doppelten oder fehlenden Stunden führen
  → [UTC](O_02_Glossar.ipynb#g-utc)-Normierung in NB2 bereinigt das

## 7. Ergebnistransfer zwischen Notebooks (../sync/transfer.json) <a id='ergebnistransfer-zwischen-notebooks-transferjson_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

### Zweck — berechnete Outputs, keine config-Redundanz

`../sync/transfer.json` enthält ausschliesslich **Werte, die ein Notebook berechnet und ein folgendes benötigt**. Was in `../sync/config.json` steht (Parameter, Schalter), wird nie nach `../sync/transfer.json` kopiert.

> **Faustregel:** `../sync/config.json` schreibt der User. `../sync/transfer.json` schreiben die Notebooks.

### Schema — vollständige Referenz

| Sektion | Geschrieben von | Gelesen von | Inhalt |
|---------|----------------|-------------|--------|
| `datenzeitraum` | **NB01** | NB02–K_99 | `start_date`, `end_date`, `n_years` — aus tatsächlich geladenen Preisdaten |
| `simulation` | **NB02** | NB04, K_00, K_05, K_10, K_99 | Spread-Kennzahlen; [ROI](O_02_Glossar.ipynb#g-roi), `net_annual`, `capex`, `payback_years` je Segment |
| `dispatch_optimierung` | **K_06** | K_00 | DA-optimal vs. reaktiv: `roi_reaktiv_pct`, `roi_da_optimal_pct`, `delta_pct` je Segment |
| `eigenverbrauch` | **K_09** | K_10 | `roi_ev_privat_pct`, `payback_ev_privat_j`, `jahresersparnis_eur` |
| `hybrid_simulation` | **K_99** | K_10 | `roi_arb/ev/hyb/hybo`, `be_*` je Segment |
| `produkt` | **K_10** | K_00 | `roi_arb_pct`, `roi_ev_pct`, `roi_kombi_pct`, `payback_kombi_j` |

### Ausführungsreihenfolge für vollständiges ../sync/transfer.json

```
NB01 → NB02 → K_06 → K_09 → K_99 → K_10
```

Jedes Notebook kann einzeln neu ausgeführt werden — es überschreibt nur seine eigene Sektion.

### Laden (alle Notebooks — robust gegen leere/fehlende Datei)

```python
import os as _os, json as _json
_tf_path = '../sync/transfer.json'
_tf = _json.loads(open(_tf_path).read() or '{}') \
      if _os.path.exists(_tf_path) and _os.path.getsize(_tf_path) > 0 else {}

# Werte lesen — immer mit Fallback:
TF_N_YEARS = _tf.get('datenzeitraum', {}).get('n_years', None)
TF_ECON    = _tf.get('simulation', {}).get('wirtschaftlichkeit', {})
```

### Schreiben (nur Notebooks die Transferwerte erzeugen)

```python
_tf = _json.loads(open(_tf_path).read() or '{}') \
      if _os.path.exists(_tf_path) and _os.path.getsize(_tf_path) > 0 else {}
_tf['simulation'] = { ... }          # nur eigene Sektion — Rest bleibt erhalten
with open(_tf_path, 'w') as _f:
    _json.dump(_tf, _f, indent=2, ensure_ascii=False)
```

### Regeln

- Nie manuell editieren — Werte stammen immer aus Notebook-Runs
- Keine `../sync/config.json`-Werte spiegeln (kein `kuer_aktiv`, keine Parameter)
- `getsize > 0`-Guard immer verwenden — verhindert Crash bei leerer Datei
- Fehlende Sektionen immer mit `.get('sektion', {})` und Fallback lesen

## 15. Sprach- & Terminologiekonvention <a id='sprach-terminologiekonvention_O_03'></a>

[↑ Inhaltsverzeichnis](#toc_O_03)

Projektsprache ist **Deutsch**. Für Fachbegriffe gilt:

> **Etablierte Fachbegriffe werden in ihrer Originalsprache verwendet —
> korrekt geschrieben, nicht eingedeutscht.**
> Denglisch (gemischte Wortteile aus zwei Sprachen) ist grundsätzlich verboten.

### Prinzip

Ein Begriff gilt als etabliert, wenn er in der Fachliteratur, in Normen oder
in der Branchenkommunikation (CH/DE/AT Energiewirtschaft) konsistent in dieser
Form verwendet wird. In diesem Fall bleibt er unverändert — er wird weder
übersetzt noch eingedeutscht.

Ist kein etablierter Fachbegriff vorhanden, wird der deutsche Begriff verwendet.

### Beispiele

| Kontext | ✅ Richtig | ❌ Falsch | Begründung |
|---|---|---|---|
| Vermarktungsstrategie | [Revenue Stacking](O_02_Glossar.ipynb#g-erloess-stacking) | Erlösstapelung | Branchenstandard |
| Regelenergie | [FCR](O_02_Glossar.ipynb#g-fcr), [aFRR](O_02_Glossar.ipynb#g-afrr) | — | [ENTSO-E](O_02_Glossar.ipynb#g-entsoe) Norm |
| Einsatzplanung | [Dispatch](O_02_Glossar.ipynb#g-dispatch) | — | Energiewirtschaft |
| Visualisierung | Heatmap | Wärmekarte | Fachvisualisierung |
| Wendepunkt | Gamechanger | **Spielchanger** | Anglizismus, aber korrekt geschrieben |
| Referenzszenario | Baseline | — | Fachsprache |
| Schwellenwert | Trigger | — | Im Investitionskontext etabliert |

### Regel für Zweifelsfälle

Wenn unklar ob ein Begriff etabliert ist: deutschen Begriff verwenden.
Neue Anglizismen nicht erfinden — insbesondere keine Wortzusammensetzungen
aus einer deutschen und einer englischen Komponente.

### 17.1 Pflicht-Notebooks (`notebooks/`)

```
5. Initialisierung  ← NUR NB01 (erstes lauffähiges Notebook):
     Abhängigkeiten prüfen · Imports · Versionen · config + transfer laden · Pfade
     Ab NB02: Imports + Laden (kein Abhängigkeits-Check, kein Versions-Print)
6. Hauptteil je Datensatz (NB01–NB04) / je Analyse mit Visuals (NB00)
7. Abschluss:
   - NB01–NB04: Zusammenfassung erzeugte Dateien + transfer.json schreiben
   - NB00: Fazit + Vorschau
8. Navigations-Links (Wiederholung von 2)
```

### 17.2 Kür-Notebooks (`kuer/`)

Jedes Kür-Notebook ist einem **einzigen Topic** gewidmet.

```
1. Titel-Zelle        # K_xx – Titel | Gruppe · Datum · Kür-Badge
2. Navigations-Links  | ← Vorheriges | ↑ Übersicht | Nächstes → |
3. Inhaltsverzeichnis
4. Einleitung
   - Motivation     (Warum ist dieses Topic relevant für Arbitrage?)
   - Forschungsfragen (konkret, beantwortbar)
   - K_00 (Business Strategy): breiter Überblick analog Business Case
5. Initialisierung    Imports · config + transfer · Pfade
   + Datenladen: Falls neue externe Quellen → hier laden (erste K die sie braucht)
6. Daten laden (neu) → Datenbereinigung → Analyse & Visualisierung
   Je Abschnitt: MD-Beschreibung → Code → Verifikation
   K_99 (Kombinierte Simulation): setzt alle vorherigen K_xx voraus
7. Fazit & Erkenntnisse
   - Beantwortung der Forschungsfragen aus Abschnitt 4
   - Einordnung in Gesamtkontext (Arbitrage-Strategie / Business Case)
8. Navigations-Links (Wiederholung von 2)
9. Abschlussblock (Datei-Checks, transfer.json schreiben falls nötig)
```

### Regeln für neue Kür-Notebooks

1. **Nummerierung:** Nächste freie Nummer zwischen K_10 und K_98 (K_99 bleibt Synthese).  
   Beispiel: K_11_Wetter_Korrelation, K_12_ML_Dispatch, …

2. **config.json:** Neue Konfigurationsparameter kommen unter `kuer → _section_<key>` + Unterkey.  
   Neuer Eintrag in `kuer_aktiv` mit `false` als Default — bei Einreichungsentscheid auf `true`.

3. **Datenquellen:** Wird eine neue externe Quelle genutzt, wird sie im neuen Notebook geladen  
   (Konvention: die erste K, die eine Quelle braucht, lädt sie). Eintrag in `dataindex.csv`.

4. **transfer.json:** Neue Outputs unter neuem Key `kuer_kNN` ablegen, nie bestehende Keys überschreiben.

5. **Navigation:** Vor K_99 einfügen: K_(N-1) next → K_N, K_N prev → K_(N-1), K_N next → K_99,  
   K_99 prev → K_N (Update in K_99 und K_(N-1) nötig).  
   K_00 prev (zirkulär) → K_99 (immer aktuell halten).

6. **Chart-Präfix:** `kuer_kNN_` gemäss Abschnitt 5d.

7. **requirements.txt:** Neue Pakete mit Versionspin hinzufügen.
